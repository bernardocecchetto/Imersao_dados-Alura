```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
dados_url = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'
dados = pd.read_csv(dados_url, compression = 'zip')
```

```python
for i in range (len(dados.columns)):
    to_list = list(dados.columns[i])
    if (np.isin(to_list[1], '-')) ==  True:
        (to_list).remove('-')
        new_column = ''.join(to_list)
        dados.rename({dados.columns[i]:new_column}, axis=1, inplace=True)
        
        

```

# DESAFIO 12: PLOTAR OS HISTOGRAMAS COM SEABORN

```python
plt.figure(figsize=(20,25))

plt.subplot(3,1,1)
sms.set(font_scale =1.8) #tamanho da fonte
ax=sms.histplot(dados.loc[:,'g0':'g771'].describe().T['mean'])
ax.set_title('Histograma dos Valores Médios de Expressão dos Genes')
ax.set(xlabel='Valor das Médias', ylabel='Incidência') 

plt.subplot(3,1,2)
sms.set(font_scale =1.8) #tamanho da fonte
ax=sms.histplot(dados.loc[:,'g0':'g771'].describe().T['min'])
ax.set_title('Histograma dos Valores Mínimos de Expressão dos Genes')
ax.set(xlabel='Valor das Mínimos', ylabel='Incidência') 

plt.subplot(3,1,3)
sms.set(font_scale =1.8) #tamanho da fonte
ax=sms.histplot(data=dados.loc[:,'g0':'g771'].describe().T['max'])
ax.set_title('Histograma dos Valores Máximos de Expressão dos Genes')
ax.set(xlabel='Valor das Máximos', ylabel='Incidência') 


```

### O que seria o boxplot? Gráfico com o objetivo de representar como está a distribuição de valores de uma variável. No gráfico abaixo, percebemos que o traço vertical está zero, representando a mediana. Ou seja, metade dos valores de g0 está acima de zero e a outra metade, abaixo de zero.

### Analisando a caixinha, ela nos representa que 25% desses valores estão entre, mais ou menos, 0 e +- 0,5.

```python
plt.figure(figsize=(20,8))

plt.subplot(1,2,1)
sms.boxplot(x = 'g0' , data=dados)

plt.subplot(1,2,2)
sms.boxplot(y = 'g0',x='tratamento' , data=dados)

```
